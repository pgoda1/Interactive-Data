# Group 1

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
diabetes = pd.read_csv('Diabetes.csv')
diabetes
```

```python
sns.scatterplot(x=diabetes['Age'], y=diabetes['Insulin'])
plt.show()
```

```python
sns.relplot(data=diabetes, x="Age", y="Insulin")
```

#The above scatterplot looks jumbled and it is a bit difficult to determine the relationship between BMI and Glucose in our dataset. So we can look at other variables in the dataset and plug them into replot to try and see what relation the variables have.

```python
sns.relplot(data=diabetes, x="Age", y="Insulin", hue="Pregnancies")
```

```python
sns.relplot(data=diabetes, x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies")
```

```python
sns.relplot(data=diabetes, x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies", sizes= (10, 50))
#Adusting the sizes of the points helped make the plot clearer and easier to differentiate the points from one another.
```

```python
sns.relplot(data=diabetes, x="Age", y="Insulin", hue="Pregnancies", size="Pregnancies", sizes= (10, 100))
plt.xlim(10, 80)  
plt.ylim(-100, 300)  

plt.show()
#To focus in more on the part of the figured we'd like to get a closer look at, we can use the .xlim and .ylim methods.
```

# Group 2

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid") 
```

```python
Salary_data = pd.read_csv("Salary Data.csv")
```

```python
Salary_data.head()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", data=Salary_data)

plt.title('Default Scatter Plot: Age vs Salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data)
plt.title( 'Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}

plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors)
plt.title( 'Scatter Plot: Age vs Salary by Education Level and Gender')

plt.show() 
```

```python
# Defining custom color palette
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
# Defining custom markers for the Gender
gender_markers = {"Female": "o", "Male": "s"}

# Creating a scatter plot with custom colors and markers
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#Changing the size of the shape and getting rid of grid lines
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}


plt.figure(figsize=(12, 8))
sns.set_style("ticks")
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')

plt.show()
```

```python
#legend Customization
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}

plt.figure(figsize=(12, 8))
sns.set_style("ticks") 
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)
#removing the legend
plt.legend(frameon=False)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
education_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gender_markers = {"Female": "o", "Male": "s"}
plt.figure(figsize=(12, 8))
sns.set_style("ticks") 
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_data, palette=education_colors, markers=gender_markers)

plt.legend(ncol=2, frameon=False, loc= "lower right" )
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks")
Bachelors=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors)

Bachelors.set_xlim(25, 50)  
Bachelors.set_ylim(50000, 250000)

plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks")
Masters=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"], palette=gender_colors)

Masters.set_xlim(25, 50)  
Masters.set_ylim(50000, 250000)

plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among Master's Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

plt.figure(figsize=(10, 8))
sns.set_style("ticks") 
phd=sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"], palette=gender_colors)

phd.set_xlim(25, 50)  
phd.set_ylim(50000, 250000)

plt.legend(frameon=False, loc= "upper left" )
plt.title("Age vs Salary Based on Gender Among PhD Degree Holders")
plt.show()
```

```python
gender_colors = {"Female": "red", "Male": "blue"}

g = sns.FacetGrid(Salary_data, col="Education Level")
g.map_dataframe(sns.scatterplot, x= "Age", y= "Salary", hue="Gender", palette=gender_colors)  

g.add_legend(title="Gender", frameon=False, loc="lower right")

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks")

sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors, ax=ax[0])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"], palette=gender_colors, ax=ax[1])
sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"], palette=gender_colors, ax=ax[2])

for axes in ax:
    axes.legend(frameon=False)


plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(14, 4))
gender_colors = {"Female": "red", "Male": "blue"}
sns.set_style("ticks") 
scatter_1 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Bachelor's"],
                            palette=gender_colors, ax=ax[0])
scatter_2 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "Master's"],
                            palette=gender_colors, ax=ax[1])
scatter_3 = sns.scatterplot(x="Age", y="Salary", hue="Gender", data=Salary_data[Salary_data["Education Level"] == "PhD"],
                            palette=gender_colors, ax=ax[2])

plt.suptitle('Scatter Plot: Age vs Salary by Education Level and Gender', ha="center")

for axes in ax:
    axes.set_xlim(25, 50)  
    axes.set_ylim(50000, 250000) 
    axes.legend(frameon=False, loc = "upper left")

plt.show()
```
