```python
import pandas as pd
import plotly.express as px
```

```python
cricket = pd.read_csv('cricket.csv')
```

```python
cricket
```

```python
fig1 = px.scatter(cricket,\
                 x = 'Innings',\
                 y = 'Runs',\
                 height = 500,\
                 title = "Cricket Player Statistics")
fig1.show()
```

```python
#Add player name and country

fig2 = px.scatter(cricket,\
                 x="Innings",\
                 y="Runs",\
                 height = 500,\
                 title = "Cricket Player Statistics",\
                 hover_name="Player",\
                 hover_data=["Country"])
fig2.show()
```

```python
# Button that changes scatterplot to bar graph

fig3 = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics" )

button1 = {'method': 'update',
          'label': 'Scatterplot',
          'args': ['type', 'scatter']
          }

button2 = {'method': 'update',
          'label': 'Bar Graph',
          'args': ['type', 'bar']
          }

fig3.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button1, button2]
    }]
)

fig3.show()
```

```python
# Button that changes All Teams to India

fig5 = px.scatter(cricket,\
                  x = "Innings",\
                  y = "Runs",\
                  hover_name = "Player",\
                  hover_data = ["Average", "Country"],\
                  height = 500,\
                  title = "Cricket Player Statistics")

df_india = cricket[cricket['Country'] == 'INDIA']

button5 = {'method': 'update',
          'label': 'All Teams',
          'args': [
              {'x': [cricket['Innings']]},
              {'y': [cricket['Runs']]}]
          }

button6 = {'method': 'update',
          'label': 'India',
          'args': [
              {'x': [df_india['Innings']]},
              {'y': [df_india['Runs']]}]
          }

fig5.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button5, button6]
    }]
)

fig5.show()
```
